# initializing parameters

take architecture of model

(how many total layers are their i am going with 2 i/p 2 hidden and 1 o/p layer)

hence 9 params will be created my initial weight is 0.1 and bias is 0

```python
def init_params(layers_dim):
  np.random.seed(3)
  params={}
  l=len(layers_dim)
  for i in range(1,l):
    params['W'+str(i)]=np.ones((layers_dim[i-1],layers_dim[i]))*0.1
    params['b'+str(i)]=np.zeros((layers_dim[i],1))
  return params
```

```python
init_params([2,2,1])
```

```python
def linear_forward(A_prev,W,b):
  Z=np.dot(W.T,A_prev)+b
  return Z
```

```python
# main forward prop

def forward_prop(X,params):
  A=X
  L=len(params)//2
  for l in range(1,L+1):
    A_prev=A
    Wl=params['W'+str(l)]
    bl=params['b'+str(l)]
    print("A"+str(l-1)+': ',A_prev)
    print("W"+str(l)+': ',Wl)
    print("b"+str(l)+': ',bl)
    print('---'*20)
    A=linear_forward(A_prev,Wl,bl)
    print("A"+str(l)+': ',A)
    print('**'*20)
  return A,A_prev
```

```python
def update_parameters(parameters,y,y_hat,A1,X):
  y = y.item() if isinstance(y, np.ndarray) else y
  y_hat = y_hat.item() if isinstance(y_hat, np.ndarray) else y_hat

  parameters['W2'][0][0] = parameters['W2'][0][0] + (0.001 * 2 * (y - y_hat)*A1[0][0])
  parameters['W2'][1][0] = parameters['W2'][1][0] + (0.001 * 2 * (y - y_hat)*A1[1][0])
  parameters['b2'][0][0] = parameters['W2'][1][0] + (0.001 * 2 * (y - y_hat))

  parameters['W1'][0][0] = parameters['W1'][0][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][0][0]*X[0][0])
  parameters['W1'][0][1] = parameters['W1'][0][1] + (0.001 * 2 * (y - y_hat)*parameters['W2'][0][0]*X[1][0])
  parameters['b1'][0][0] = parameters['b1'][0][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][0][0])

  parameters['W1'][1][0] = parameters['W1'][1][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][1][0]*X[0][0])
  parameters['W1'][1][1] = parameters['W1'][1][1] + (0.001 * 2 * (y - y_hat)*parameters['W2'][1][0]*X[1][0])
  parameters['b1'][1][0] = parameters['b1'][1][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][1][0])
```

```python
update_parameters(parameters,y,y_hat,A1,X)
```

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
model =Sequential()
model.add(Dense(2,activation='linear',input_dim=2))
model.add(Dense(1,activation='linear'))
```

```python
model.summary()
```

```python
model.get_weights()
```

```python
# making the weights similar to mine

new_weights=[np.array([[0.1,0.1],[0.1,0.1]]),
             np.array([0.,0.]),
             np.array([[0.1],[0.1]]),
             np.array([0.])]
```

```python
model.set_weights(new_weights)
```

```python
model.get_weights()
```

```python
optimizer=keras.optimizers.Adam(learning_rate=0.001)
model.compile(loss='mean_squared_error',optimizer=optimizer)
```

```python
model.fit(df.iloc[:,0:-1].values,df['lpa'].values,epochs=10,verbose=1,batch_size=1)
```
